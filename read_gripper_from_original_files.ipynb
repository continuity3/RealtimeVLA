```python
# 更准确的对比：模拟转换脚本的完整流程
print("=" * 80)
print("准确对比：模拟转换脚本的完整处理流程")
print("=" * 80)

if len(all_gripper_data) > 0 and 'gripper_states_from_dataset' in locals():
    # 获取第一个文件的原始夹爪数据
    first_hdf5_path = hdf5_files[0]
    original_gripper = all_gripper_data[0]['values']
    
    # 模拟转换脚本的处理流程
    with h5py.File(first_hdf5_path, "r") as f:
        # 1. 检查是否有 valid 标记
        valid = None
        if "valid" in f:
            if "_joint_states" in f["valid"]:
                valid_joint = f["valid/_joint_states"][:]  # (T,)
            else:
                valid_joint = None
            
            if "_camera_camera_color_image_raw" in f["valid"]:
                valid_image = f["valid/_camera_camera_color_image_raw"][:]  # (T,)
            else:
                valid_image = None
            
            if valid_joint is not None and valid_image is not None:
                valid = valid_joint & valid_image
            elif valid_joint is not None:
                valid = valid_joint
            elif valid_image is not None:
                valid = valid_image
        
        # 2. 计算 valid_indices（模拟转换脚本的逻辑）
        min_length = len(original_gripper)
        if valid is not None:
            valid_indices = np.where(valid[:min_length])[0]
            print(f"\n✅ 找到 valid 标记，过滤后有效索引数量: {len(valid_indices)} / {min_length}")
            print(f"   有效索引范围: [{valid_indices[0] if len(valid_indices) > 0 else 'N/A'}, {valid_indices[-1] if len(valid_indices) > 0 else 'N/A'}]")
        else:
            valid_indices = np.arange(min_length)
            print(f"\n⚠️  未找到 valid 标记，使用所有数据: {len(valid_indices)} 个点")
        
        # 3. 先过滤有效索引（模拟转换脚本第357行）
        filtered_gripper = original_gripper[valid_indices]
        print(f"   过滤后数据点数量: {len(filtered_gripper)}")
        
        # 4. 下采样（模拟转换脚本第377行和第387行）
        downsample_factor = 4
        downsampled_indices = np.arange(0, len(filtered_gripper), downsample_factor)
        downsampled_gripper = filtered_gripper[downsampled_indices]
        print(f"   下采样后数据点数量: {len(downsampled_gripper)} (每 {downsample_factor} 个取1个)")
        
        # 5. 处理 NaN（模拟转换脚本第439行的 handle_nan_values）
        def handle_nan_values_simple(data):
            """简化版的 handle_nan_values，只处理1D数组"""
            data_clean = data.copy()
            for i in range(len(data_clean)):
                if np.isnan(data_clean[i]):
                    if i == 0:
                        # 第一个：使用下一个非 NaN 值
                        for j in range(1, len(data_clean)):
                            if not np.isnan(data_clean[j]):
                                data_clean[i] = data_clean[j]
                                break
                    elif i == len(data_clean) - 1:
                        # 最后一个：使用上一个非 NaN 值
                        for j in range(len(data_clean) - 2, -1, -1):
                            if not np.isnan(data_clean[j]):
                                data_clean[i] = data_clean[j]
                                break
                    else:
                        # 中间：使用前一个非 NaN 值
                        for j in range(i - 1, -1, -1):
                            if not np.isnan(data_clean[j]):
                                data_clean[i] = data_clean[j]
                                break
            return data_clean
        
        processed_gripper = handle_nan_values_simple(downsampled_gripper)
        print(f"   处理 NaN 后数据点数量: {len(processed_gripper)}")
        
        # 6. 对比
        print(f"\n转换后数据集中的夹爪状态数量: {len(gripper_states_from_dataset)}")
        print(f"模拟处理后的夹爪状态数量: {len(processed_gripper)}")
        
        # 对比值
        min_len = min(len(processed_gripper), len(gripper_states_from_dataset))
        print(f"\n前 {min_len} 个值对比:")
        print(f"{'索引':>6} | {'模拟处理':>12} | {'转换后数据集':>12} | {'匹配':>6} | {'差异':>12}")
        print("-" * 60)
        
        match_count = 0
        for i in range(min_len):
            sim_val = processed_gripper[i]
            conv_val = gripper_states_from_dataset[i]
            
            # 检查是否匹配（考虑浮点数精度）
            is_match = np.isclose(sim_val, conv_val, rtol=1e-5, atol=1e-6) or \
                       (np.isnan(sim_val) and np.isnan(conv_val))
            
            if is_match:
                match_count += 1
            
            match_str = "✅" if is_match else "❌"
            sim_str = f"{sim_val:.6f}" if not np.isnan(sim_val) else "NaN"
            conv_str = f"{conv_val:.6f}" if not np.isnan(conv_val) else "NaN"
            diff_str = f"{abs(sim_val - conv_val):.6f}" if not (np.isnan(sim_val) or np.isnan(conv_val)) else "N/A"
            
            print(f"{i:6d} | {sim_str:>12} | {conv_str:>12} | {match_str:>6} | {diff_str:>12}")
        
        print(f"\n匹配情况: {match_count}/{min_len} 个值匹配 ({match_count/min_len*100:.1f}%)")
        
        # 如果还有不匹配的值，显示详细信息
        if match_count < min_len:
            print(f"\n不匹配的值:")
            for i in range(min_len):
                sim_val = processed_gripper[i]
                conv_val = gripper_states_from_dataset[i]
                is_match = np.isclose(sim_val, conv_val, rtol=1e-5, atol=1e-6) or \
                           (np.isnan(sim_val) and np.isnan(conv_val))
                if not is_match:
                    print(f"  索引 {i}: 模拟={sim_val:.6f}, 转换后={conv_val:.6f}, 差异={abs(sim_val - conv_val):.6f}")
        
else:
    print("❌ 无法对比：缺少原始数据或转换后数据")
```

```python
# 读取转换后数据集中对应 rosbag2_2026_01_09-21_24_48_0.h5 的 state 和 action
import pyarrow.parquet as pq
from pathlib import Path

print("=" * 80)
print("读取转换后数据集中对应 rosbag2_2026_01_09-21_24_48_0.h5 的 state 和 action")
print("=" * 80)

# 目标文件名
target_file = "rosbag2_2026_01_09-21_24_48_0.h5"

# 数据集路径
lerobot_cache = Path.home() / ".cache" / "huggingface" / "lerobot"
dataset_path = lerobot_cache / "your_hf_username" / "pick_blue_bottle_libero_downsample4x"

if dataset_path.exists():
    # 1. 找到目标文件在原始文件列表中的索引
    original_data_dir = Path("pick_blue_bottle")
    hdf5_files = sorted(
        list(original_data_dir.glob("**/*.h5")) + list(original_data_dir.glob("**/*.hdf5"))
    )
    
    # 找到目标文件的索引
    target_index = None
    for i, hdf5_path in enumerate(hdf5_files):
        if hdf5_path.name == target_file:
            target_index = i
            break
    
    if target_index is None:
        print(f"❌ 未找到文件: {target_file}")
    else:
        print(f"✅ 找到文件: {target_file}")
        print(f"   在排序后的文件列表中索引: {target_index}")
        print(f"   对应 episode: episode_{target_index:06d}.parquet")
        
        # 2. 读取对应的 parquet 文件
        parquet_files = sorted(list(dataset_path.glob("**/*.parquet")))
        if target_index < len(parquet_files):
            target_parquet = parquet_files[target_index]
            print(f"\n读取文件: {target_parquet.name}")
            
            table = pq.read_table(target_parquet)
            print(f"  列: {table.column_names}")
            print(f"  行数: {len(table)}")
            
            # 读取 state 和 action
            if 'state' in table.column_names and 'actions' in table.column_names:
                states = table['state'].to_numpy()
                actions = table['actions'].to_numpy()
                
                # 处理 state（可能是 object 类型）
                if states.dtype == object:
                    states_2d = np.array([s if isinstance(s, np.ndarray) else np.array(s) for s in states])
                else:
                    states_2d = states
                
                # 处理 action（可能是 object 类型）
                if actions.dtype == object:
                    actions_2d = np.array([a if isinstance(a, np.ndarray) else np.array(a) for a in actions])
                else:
                    actions_2d = actions
                
                print(f"\nState 数据:")
                print(f"  Shape: {states_2d.shape}")
                print(f"  Dtype: {states_2d.dtype}")
                print(f"  维度说明: 8维 (7个关节 + 1个夹爪)")
                
                print(f"\nAction 数据:")
                print(f"  Shape: {actions_2d.shape}")
                print(f"  Dtype: {actions_2d.dtype}")
                print(f"  维度说明: 7维 (6维 EEF action + 1维夹爪 action)")
                
                # 打印所有 state 和 action
                print(f"\n" + "=" * 80)
                print(f"完整的 State 和 Action 数据（共 {len(states_2d)} 个时间步）")
                print("=" * 80)
                
                for i in range(len(states_2d)):
                    state = states_2d[i]
                    action = actions_2d[i]
                    
                    # 分解 state：7个关节 + 1个夹爪
                    joints = state[:7]
                    gripper_state = state[7]
                    
                    # 分解 action：6维 EEF action + 1维夹爪 action
                    eef_action = action[:6]  # [dx, dy, dz, drx, dry, drz]
                    gripper_action = action[6]
                    
                    print(f"\n时间步 [{i:3d}]:")
                    print(f"  State (8维):")
                    print(f"    关节位置 (7维): {joints}")
                    print(f"    夹爪状态 (1维): {gripper_state:.6f}")
                    print(f"  Action (7维):")
                    print(f"    EEF action (6维): [dx={eef_action[0]:8.6f}, dy={eef_action[1]:8.6f}, dz={eef_action[2]:8.6f}, "
                          f"drx={eef_action[3]:8.6f}, dry={eef_action[4]:8.6f}, drz={eef_action[5]:8.6f}]")
                    print(f"    夹爪 action (1维): {gripper_action:.6f}")
                
                # 统计信息
                print(f"\n" + "=" * 80)
                print("统计信息")
                print("=" * 80)
                
                print(f"\nState 统计:")
                print(f"  关节位置 (前7维):")
                for j in range(7):
                    joint_values = states_2d[:, j]
                    print(f"    关节 {j+1}: min={np.nanmin(joint_values):.6f}, max={np.nanmax(joint_values):.6f}, "
                          f"mean={np.nanmean(joint_values):.6f}, std={np.nanstd(joint_values):.6f}")
                print(f"  夹爪状态 (第8维):")
                gripper_states = states_2d[:, 7]
                print(f"    min={np.nanmin(gripper_states):.6f}, max={np.nanmax(gripper_states):.6f}, "
                      f"mean={np.nanmean(gripper_states):.6f}, std={np.nanstd(gripper_states):.6f}")
                
                print(f"\nAction 统计:")
                print(f"  EEF action (前6维):")
                action_names = ["dx", "dy", "dz", "drx", "dry", "drz"]
                for j in range(6):
                    action_values = actions_2d[:, j]
                    print(f"    {action_names[j]}: min={np.nanmin(action_values):.6f}, max={np.nanmax(action_values):.6f}, "
                          f"mean={np.nanmean(action_values):.6f}, std={np.nanstd(action_values):.6f}")
                print(f"  夹爪 action (第7维):")
                gripper_actions = actions_2d[:, 6]
                print(f"    min={np.nanmin(gripper_actions):.6f}, max={np.nanmax(gripper_actions):.6f}, "
                      f"mean={np.nanmean(gripper_actions):.6f}, std={np.nanstd(gripper_actions):.6f}")
                
                # 保存 action 为 npy 文件
                action_npy_file = Path(f"actions_{target_file.replace('.h5', '')}.npy")
                np.save(action_npy_file, actions_2d)
                print(f"\n✅ Action 数据已保存到: {action_npy_file.absolute()}")
                print(f"   Shape: {actions_2d.shape}")
                print(f"   Dtype: {actions_2d.dtype}")
                
            else:
                print("❌ 表中没有 'state' 或 'actions' 列")
        else:
            print(f"❌ Episode 索引 {target_index} 超出范围（共 {len(parquet_files)} 个 episode）")
else:
    print(f"❌ 数据集路径不存在: {dataset_path}")
```
